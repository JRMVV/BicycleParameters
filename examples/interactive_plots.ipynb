```python
import numpy as np
import matplotlib.pyplot as plt
import bicycleparameters as bp
bicycle = bp.Bicycle('Benchmark', pathToData='data')
```

```python
%matplotlib inline
```

```python
speeds = np.linspace(0., 10., num=100)
```

```python
def change_bicycle(c, lam, rF):
    bicycle.parameters['Benchmark']['c'] = c
    bicycle.parameters['Benchmark']['lam'] = lam
    bicycle.parameters['Benchmark']['rF'] = rF
    bicycle.plot_bicycle_geometry()
    bicycle.plot_eigenvalues_vs_speed(speeds)
    plt.ylim((-10, 10))
```

```python
from IPython.html.widgets import interact, interactive
from IPython.display import clear_output, display, HTML
w = interactive(change_bicycle, c=(-0.1, 0.1, 0.01), lam=(0.27, 0.43, 0.01), rF=(0.0, 0.5, 0.01))
display(w)
```
